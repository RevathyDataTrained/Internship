```python
#some feature are not normalized we need to normalize using statndard scalar
x=df.drop(columns="W")
y=df["W"]
```

```python
x.skew()
```

```python
from scipy import stats
```

```python
x["R"]=stats.boxcox(x["R"])[0]
x["E"]=stats.boxcox(x["E"])[0]
```

```python
x.skew()
#skewness removed
```

```python
#check for outliers
column=x.columns.values
plt.figure(figsize=(30,90))
for i in range(len(column)):
    plt.subplot(17,30,i+1)
    sns.boxenplot(data=x[column[i]])
```

```python
#remove outliers
from scipy.stats import zscore
```

```python
z=np.abs(zscore(x))
```

```python
z
```

```python
x=x[(z<3).all(axis=1)]
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
from  sklearn.preprocessing import StandardScaler
lr=StandardScaler()
x=lr.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
LR=LinearRegression()
LR=LR.fit(x_train,y_train)
y_Ptrain=LR.predict(x_train)
y_Ptest=LR.predict(x_test)
score1=r2_score(y_train,y_Ptrain)
score2=mean_absolute_error(y_train,y_Ptrain)
score3=mean_squared_error(y_train,y_Ptrain)
print(score1)
print(score2)
print(score3)
```

```python
from sklearn.model_selection import cross_val_score,GridSearchCV
trainscore=r2_score(y_Ptrain,y_train)
testscore=r2_score(y_test,y_Ptest)
for i in range(2,10):
    crossval=cross_val_score(LR,x_train,y_train,cv=i)
    cvmean=crossval.mean()
    print("At cv= {}, cross mean is {},trainscore={} and testscore={}".format(i,cvmean,trainscore,testscore))
```

```python
from sklearn.linear_model import Lasso
ls=Lasso()
params = {'alpha': [0.001, 0.01, 0.1, 1, 10],'random_state':list(range(1,10))}
lsm = GridSearchCV(ls, params)

lsm.fit(x_train, y_train)

print(lsm.best_params_)
print(lsm.best_score_)
```

```python
ls=Lasso(alpha=0.1,random_state=1)
lsm.fit(x_train, y_train)
pred=lsm.predict(x_test)
score=r2_score(y_test,pred)
score
```

```python
crossval=cross_val_score(lsm,x_train,y_train,cv=2)
cvmean=crossval.mean()
cvmean
```

```python
from sklearn.ensemble import RandomForestRegressor
import warnings
warnings.filterwarnings('ignore')
```

```python
rm=RandomForestRegressor()
para={'criterion':['mse','mae'],'max_features':["auto","sqrt","log2"]}
rmG=GridSearchCV(rm,para)
rmG.fit(x_train,y_train)
rmG.best_params_
```

```python
rm=RandomForestRegressor(criterion="mse", max_features='auto')
rmG.fit(x_train, y_train)
pred=rmG.predict(x_test)
score=r2_score(y_test,pred)
score
```

```python
crossval=cross_val_score(rmG,x_train,y_train,cv=2)
cvmean=crossval.mean()
cvmean
```

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
dt.fit(x_train, y_train)
pred=dt.predict(x_test)
score=r2_score(y_test,pred)
score
```

```python
crossval=cross_val_score(dt,x_train,y_train,cv=2)
cvmean=crossval.mean()
cvmean
```

```python
from sklearn.neighbors import KNeighborsRegressor
kn=KNeighborsRegressor()
kn.fit(x_train, y_train)
pred=kn.predict(x_test)
score=r2_score(y_test,pred)
score
```
