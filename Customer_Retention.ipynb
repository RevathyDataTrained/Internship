```python
#find the values counts for all the columns
columns=df.columns
for i in columns:
    print("------------------"+i+"-------------------")
    print(df[i].value_counts())
```

```python
#find the unique values  for all the columns
columns=df.columns
for i in columns:
    print("------------------"+i+"-------------------")
    print(df[i].unique())
```

```python
#find length of the unique values
#each columns have less number of unique values
columns=df.columns
for i in columns:
  
    print("{} contains {} values ".format(i,len(df[i].unique())))
```

```python
plt.figure(figsize=(30,30))
fig, axes = plt.subplots(9, 9, figsize=(12, 7), sharey=True)
for i in columns:
    for j in range(0,9):
        for k in range(0,9):
            
            sns.barplot(df[i].value_counts().keys(), df[i].value_counts(), ax=axes[j,k])
```

```python
#encoding the column
from sklearn.preprocessing import LabelEncoder
lbl=LabelEncoder()
for i in columns:
    df[i]=lbl.fit_transform(df[i])
```
