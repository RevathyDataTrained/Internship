```python
#from the above graph. there is high probability of multicolinearity problrm,since shuked weight,viscera weight having less relation so i am going to drop these columns
abalone=abalone.drop(columns=["Shucked weight","Viscera weight"])
```

```python
abalone["Age"]=abalone["Rings"]+1.5
```

```python
plt.figure(figsize=(15,5))
sns.countplot(abalone["Age"])
#from the graph it clearly says large values between 8 and 15. so value less then 8 is zero (young)and above is 1(old)
```

```python
def agevalidation(x):
    if x>8:
        return 1
    else:
        return 0
    
```

```python
abalone["Age"]=abalone["Age"].apply(agevalidation)
```

```python
#remove outliers
from scipy.stats import zscore
```

```python
z=np.abs(zscore(abalone))
```

```python
dfnew=abalone[(z<3).all(axis=1)]
```

```python
dfnew.shape
```

```python
#standarize
from sklearn.preprocessing import StandardScaler
x=dfnew.drop(columns=["Rings","Age"])
Y=dfnew["Age"]
```

```python
standard=StandardScaler()
X=standard.fit(x)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
x_train,x_test,y_train,y_test=train_test_split(x,Y,random_state=47,test_size=0.25)
clf=DecisionTreeClassifier()
clf.fit(x_train,y_train)
ypred=clf.predict(x_test)
score=accuracy_score(y_test,ypred)
score

```

```python
from sklearn.svm import SVC
svc_model=SVC()
svc_model.fit(x_train,y_train)
ypred=svc_model.predict(x_test)
ascore=accuracy_score(y_test,ypred) 
ascore
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier() 
rf.fit(x_train,y_train)
ypred=rf.predict(x_test)
ascore=accuracy_score(y_test,ypred) 
ascore
```

```python

from sklearn.model_selection import GridSearchCV
params = {'C': [0.001, 0.01, 0.1, 1, 10],'gamma':[0.001, 0.01, 0.1, 1]}
svcClf = GridSearchCV(svc_model, param_grid = params, scoring = 'accuracy', cv = 10)

svcClf.fit(x_train, y_train)

print(svcClf.best_params_)
print(svcClf.best_score_)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
ypred=lr.predict(x_test)
ascore=accuracy_score(y_test,ypred) 
ascore
```

```python
#Among five model Logistic Regression is the best model
```
