```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hr=pd.read_csv("https://raw.githubusercontent.com/RevathyDataTrained/Internship/main/WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
plt.figure(figsize=(60,60))
sns.heatmap(hr.corr(),annot=True)
```

```python
hr.Attrition.value_counts()
```

```python
hr["Attrition"]=hr["Attrition"].replace(["Yes","No"],[1,0])
```

```python
hr.Attrition.value_counts()
#i have replaced yes as 1 and no as 0
```

```python
hr[hr.columns[0:]].corr()["Attrition"][:]
#shows relationship with the other columns in dataset
```

```python
#lets remove unwanted columns from the dataset
hr=hr.drop(columns=["EmployeeCount","EmployeeCount","StandardHours","HourlyRate","MonthlyRate","YearsSinceLastPromotion","Over18",'MaritalStatus'])
```

```python
hr.shape
```

```python

hr.info()
```

```python
#there are 7 categoral value. we need to convert to dummy variables
```

```python
dummy_BusinessTravel=pd.get_dummies(hr["JobRole"],prefix="BusinessTravel")
dummy_EducationField=pd.get_dummies(hr["EducationField"],prefix="EducationField")
dummy_Department=pd.get_dummies(hr["Department"],prefix="Department")
dummy_Gender=pd.get_dummies(hr["Gender"],prefix="Gender")
dummy_PercentSalaryHike=pd.get_dummies(hr["PercentSalaryHike"],prefix="PercentSalaryHike")
dummy_DailyRate=pd.get_dummies(hr["DailyRate"],prefix="DailyRate")
dummy_JobRole=pd.get_dummies(hr["JobRole"],prefix="JobRole")
dummy_OverTime=pd.get_dummies(hr["OverTime"],prefix="OverTime")


```

```python
hr["OverTime"].value_counts()
```

```python
x=hr.drop(columns=["JobRole","BusinessTravel",'Department','Gender','PercentSalaryHike','DailyRate','Attrition','EducationField','OverTime'])
x.info()
```

```python
for i in x.columns:
    sns.distplot(x[i])
    plt.show()
#datas are not normally distributed    
```

```python
x.skew()
```

```python
x=pd.concat([x,dummy_BusinessTravel,dummy_EducationField,dummy_Department,dummy_Gender,dummy_PercentSalaryHike,dummy_DailyRate,dummy_JobRole,dummy_OverTime],axis=1)
```

```python
y=hr["Attrition"]
```

```python
x.shape
```

```python
from sklearn.preprocessing import StandardScaler
ls=StandardScaler()
x=ls.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=40)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve,confusion_matrix,roc_auc_score,accuracy_score
```

```python
lr=LogisticRegression()
lr.fit(x,y)
y_pred=lr.predict(x_test)
acuracy=accuracy_score(y_test,y_pred)
```

```python
acuracy
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=45,test_size=0.20,)
lr=LogisticRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
score=accuracy_score(y_test,y_pred)
```

```python
from sklearn import metrics
confusion_matrix=metrics.confusion_matrix(y_test,y_pred)
display=metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix)
display.plot()
plt.show()
    
```

```python
print(metrics.classification_report(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import  GridSearchCV
clf=DecisionTreeClassifier()
grid_parameter={
    "criterion":["gini","entropy"],
    "max_depth":range(10,15),
    "min_samples_leaf":range(2,10),
    "min_samples_split":range(3,10),
    "max_leaf_nodes":range(2,4)
}
```

```python
grid_search=GridSearchCV(estimator=clf,param_grid=grid_parameter,cv=10,n_jobs=-1)
grid_search.fit(x_train,y_train)
```

```python
best_paramter=grid_search.best_params_
print(best_paramter)
```

```python
clf=DecisionTreeClassifier(criterion= 'entropy', max_depth= 10, max_leaf_nodes= 2, min_samples_leaf= 2, min_samples_split= 3)
clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
score=accuracy_score(y_test,y_pred)
score
```

```python
confusion_matrix=metrics.confusion_matrix(y_test,y_pred)
display=metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix)
display.plot()
plt.show()
```

```python
print(metrics.classification_report(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
svc_model=SVC()
params = {'C': [0.001, 0.01, 0.1, 1, 10],'gamma':[0.001, 0.01, 0.1, 1]}
svcClf = GridSearchCV(svc_model, param_grid = params, scoring = 'accuracy', cv = 10)

svcClf.fit(x_train, y_train)

print(svcClf.best_params_)
print(svcClf.best_score_)
```

```python
svc_model=SVC()
params = {'C':[1] , 'gamma': [0.1]}
svcClf = GridSearchCV(svc_model, param_grid = params, scoring = 'accuracy', cv = 10)

svcClf.fit(x_train, y_train)
```

```python
y_pred=svcClf.predict(x_test)
score=accuracy_score(y_test,y_pred)
score
```
